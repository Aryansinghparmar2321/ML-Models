```python
x_train = x[:60000]
y_train = y[:60000]

x_test = x[60000:]
y_test = y[60000:]

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
y_test_5 = (y_test=='5')
y_train_5 = (y_train=='5')
```

```python
y_train_5[:5]
```

```python
#SDG
from sklearn.linear_model import SGDClassifier
```

```python
model = SGDClassifier(random_state=42)
```

```python
#train
model.fit(x_train, y_train_5)
```

```python
model.predict([x[-1]])
```

```python
y[-1]
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_train_5, model.predict(x_train)))
```

```python
print(confusion_matrix(y_test_5, model.predict(x_test)))
```
